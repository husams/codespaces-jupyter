```python

import pandas as pd
import numpy as np
from scipy.stats import norm

# Parameters
mu = np.array([97.6, 98.1, 98.6])
sigma = np.array([0.5, 1.0])

# Create all possible (mu, sigma) pairs
theta = pd.DataFrame(np.array(np.meshgrid(mu, sigma)).T.reshape(-1, 2), columns=["mu", "sigma"])

# Prior
prior_mu = np.array([0.20, 0.30, 0.50])
prior_sigma = np.array([0.25, 0.75])
prior = np.prod(np.array(np.meshgrid(prior_mu, prior_sigma)).T.reshape(-1, 2), axis=1)
prior /= np.sum(prior)

# Data
y = np.array([97.9, 97.5])

# Likelihood
likelihood = norm.pdf(y[0], loc=theta['mu'], scale=theta['sigma']) * norm.pdf(y[1], loc=theta['mu'], scale=theta['sigma'])

# Posterior
product = likelihood * prior
posterior = product / np.sum(product)

# Bayes table
bayes_table = pd.DataFrame({
    'mu': theta['mu'],
    'sigma': theta['sigma'],
    'prior': prior,
    'likelihood': likelihood,
    'product': product,
    'posterior': posterior
})

# Display the table with appropriate formatting
bayes_table = bayes_table.round(4)
bayes_table
```
